```python
import json
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.rc('xtick', labelsize=16)
plt.rc('ytick', labelsize=16)
```

```python
with open("../results/ffnn_second_run_results.json", "r") as infile:
    ffnn = json.load(infile)
with open("../results/lstm_results.json", "r") as infile:
    lstm = json.load(infile)
with open("../results/roc_auc_ffnn.json", "r") as infile:
    ffnn_roc = json.load(infile)
with open("../results/roc_auc_lstm.json", "r") as infile:
    lstm_roc = json.load(infile)
```

```python
def getKey(dct, key):
    lst = dct[key]
    lst = [float(item) for item in lst]
    return lst
```

```python
x = list(range(1, len(ffnn["accuracy"])+1))
plt.figure(figsize=(12, 8))
plt.plot(x, getKey(ffnn, "val_accuracy"), "-", label="Validation")
plt.plot(x, getKey(ffnn, "accuracy"), "-", label="Training")
plt.legend(loc=0, fontsize=16)
_ = plt.xlabel("Number of epochs", fontsize=18)
_ = plt.ylabel("Accuracy", fontsize=18)
plt.savefig("../fig/ffnn_accuracy.png")
```

```python
x = list(range(1, len(ffnn["accuracy"])+1))
plt.figure(figsize=(12, 8))
plt.plot(x, getKey(ffnn, "val_loss"), "-", label="Validation")
plt.plot(x, getKey(ffnn, "loss"), "-", label="Training")
plt.legend(loc=0, fontsize=16)
_ = plt.xlabel("Number of epochs", fontsize=18)
_ = plt.ylabel("Loss", fontsize=18)
plt.savefig("../fig/ffnn_loss.png")
```

```python
x = list(range(1, len(lstm["accuracy"])+1))
plt.figure(figsize=(12, 8))
plt.plot(x, getKey(lstm, "val_accuracy"), "-", label="Validation")
plt.plot(x, getKey(lstm, "accuracy"), "-", label="Training")
plt.legend(loc=0, fontsize=16)
_ = plt.xlabel("Number of epochs", fontsize=18)
_ = plt.ylabel("Accuracy", fontsize=18)
plt.savefig("../fig/lstm_accuracy.png")
```

```python
x = list(range(1, len(lstm["accuracy"])+1))
plt.figure(figsize=(12, 8))
plt.plot(x, getKey(lstm, "val_loss"), "-", label="Validation")
plt.plot(x, getKey(lstm, "loss"), "-", label="Training")
plt.legend(loc=0, fontsize=16)
_ = plt.xlabel("Number of epochs", fontsize=18)
_ = plt.ylabel("Loss", fontsize=18)
plt.savefig("../fig/lstm_loss.png")
```

```python
plt.figure(figsize=(12, 8))
x = np.arange(0, 1.1, 0.1)
plt.plot(getKey(ffnn_roc, "fpr"), getKey(ffnn_roc, "tpr"), "-", label="Feed Forward Neural Network (AUC = 0.95)")
plt.plot(getKey(lstm_roc, "fpr"), getKey(lstm_roc, "tpr"), "-", label="LSTM Neural Network (AUC = 0.97)")
plt.plot(x, x, "--")
plt.legend(loc=0, fontsize=16)
_ = plt.xlabel("False Positive Rate", fontsize=18)
_ = plt.ylabel("True Positive Rate", fontsize=18)
plt.savefig("../fig/roc_ffnn_lstm.png")
```
